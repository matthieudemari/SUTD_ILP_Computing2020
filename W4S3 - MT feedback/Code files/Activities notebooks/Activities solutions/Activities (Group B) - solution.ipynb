```python
def is_prime(number):
    
    # Cover for 1 as particular case
    if number == 1:
        return False
    
    # Check for all numbers in [2, number]
    # If one is a divisor, the number is not prime
    for divisor in range(2, int(number**0.5) + 1):
        if(number % divisor == 0):
            return False
    
    return True

print(is_prime(1))
print(is_prime(3))
print(is_prime(6))
```
