```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file = "./treated/Activities_A.xlsx"
data = pd.read_excel(file).to_numpy()

grades_act_A = [line[-1] for line in data]
plt.subplots()
plt.hist(grades_act_A, bins = 30)
plt.xlabel("Grade")
plt.ylabel("Students number")
plt.title("Average Activity grade distribution (Group A)")
plt.show()

grades_per_q = [[data[i, index] for i in range(data.shape[0])] for index in range(2, 7)]
avg_grades_per_q = [np.mean(sub)/np.max(sub) for sub in grades_per_q]

plt.subplots()
plt.bar(range(1,6), avg_grades_per_q)
plt.xlabel("Question")
plt.ylabel("Average score on question (in percents)")
plt.title("Average questions score distribution (Group A)")
plt.show()
```

```python
file = "./treated/Final_A.xlsx"
data = pd.read_excel(file).to_numpy()

grades_act_A = [line[-1] for line in data]
plt.subplots()
plt.hist(grades_act_A, bins = 20)
plt.xlabel("Grade")
plt.ylabel("Students number")
plt.title("Average Midterm grade distribution (Group A)")
plt.show()

avg = round(np.mean(grades_act_A), 1)
var = round(np.std(grades_act_A), 1)
print("Average Midterm Score (Group A): {} +/- {}".format(avg, var))
```
